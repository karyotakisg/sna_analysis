```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/nodes.csv')
distribution = df.groupby('bridgingcoefficient').size().reset_index(name='count')
distribution.sort_values(by='count',ascending=False, inplace=True)
```

```python
plt.xlabel('Bridging Coefficient')
plt.ylabel('Frequency of Nodes') 
sns.histplot(data=df, x='bridgingcoefficient', binwidth=500, color='skyblue')
plt.savefig('../figures/bridging_coefficient_hist.png')
```
