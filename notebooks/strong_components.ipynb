```python
import pandas as pd 
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
edges = pd.read_csv('../data/edges.csv')
source = edges['Source']
target = edges['Target']
nodes = pd.read_csv('../data/nodes.csv')
nodes = nodes['Id']
G = nx.DiGraph()
G.add_edges_from(zip(source, target))
G.add_nodes_from(nodes)
```

```python
nx.strongly_connected_components(G)
[len(c) for c in sorted(nx.strongly_connected_components(G), key=len, reverse=True)]
```

```python
componets = [len(c) for c in sorted(nx.strongly_connected_components(G), key=len, reverse=True)]
plt.yscale('log')
plt.hist(componets, bins=100, color='orange')
plt.xlabel('Size of the component')
plt.ylabel('Log Number of components')
plt.savefig('../figures/strongly_connected_components.png')
```
