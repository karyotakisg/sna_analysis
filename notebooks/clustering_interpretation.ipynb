```python
df = pd.read_csv('data/nodes.csv')
```

```python
import pandas as pd
import re

# Define categories with regex
categories = {
    'Academia': r'\b(professor|researcher|academic|university|science|phd|lecturer|education|scholar|student|postdoc|instructor|faculty|alumni|associate professor)\b',
    'Technology': r'\b(engineer|developer|tech|technology|coding|programmer|AI|artificial intelligence|machine learning|data|cloud|security|software|open source|blockchain|digital|cybersecurity|automation|dev)\b',
    'Creative': r'\b(author|writer|artist|musician|podcast|book|art|editor|creator|photographer|design|designer|blog|media|substack|creator)\b',
    'Policy': r'\b(policy|politics|advocate|activist|democracy|law|climate|nonprofit|public health|sustainability|environment|human rights|justice|civil rights)\b',
    'Business': r'\b(founder|director|CEO|entrepreneur|business|startup|management|consultant|product|leader|strategy|executive|marketing|operations|team)\b',
    'Miscellaneous': r'\b(love|enthusiast|mom|dad|dog|life|view|good|newsletter|community|DM|opinion|thing|former)\b'
}

# Define modularity classes
modularity_classes = range(13)

# Create a dictionary to store results
clusters = {}

# Loop through each modularity class
for i in modularity_classes:
    cluster_results = {}
    print(f"Community {i}:")
    
    # Filter the DataFrame for the current modularity class
    new_df = df[df['modularity_class'] == i]
    
    # Apply regex for each category and count matches
    for key, regex in categories.items():
        new_df[key] = new_df['Description'].str.contains(regex, flags=re.IGNORECASE, na=False)
        cluster_results[key] = new_df[key].value_counts().to_dict()  # Store True/False counts
    
    # Store results for the current modularity class
    clusters[i] = cluster_results
```

```python
for key, categories in clusters.items():
    # Build a list of tuples: (category_name, count)
    cat_list = [(cat_name, cat_data[1]) for cat_name, cat_data in categories.items()]
    
    # Sort the list in descending order by count
    sorted_categories = sorted(cat_list, key=lambda x: x[1], reverse=True)
    
    # Get the top two categories
    top_two = sorted_categories[:2]
    
    # Print the result for this cluster
    print(f"Community {key}: Top Categories: {top_two}")
```
