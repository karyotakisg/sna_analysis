```python
import pandas as pd
import networkx as nx
```

```python
nodes_df = pd.read_csv('../data/nodes.csv')
edges_df = pd.read_csv('../data/edges.csv')
G = nx.Graph()
source = edges_df['Source']
target = edges_df['Target']
nodes = nodes_df['Id']
description = nodes_df['Description']
G.add_edges_from(zip(source, target))
G.add_nodes_from(nodes)
nx.degree_assortativity_coefficient(G)
```

```python
gender = pd.read_csv('../data/classification.csv')
nodes_df = pd.concat([nodes_df,gender],axis=1)
gender_dict = nodes_df.set_index('Id')['Gender'].to_dict()
gender_dict
```

```python
nx.set_node_attributes(G, values=gender_dict, name='Gender')
nx.attribute_assortativity_coefficient(G,'Gender')
```

```python
nodes_df['Gender'] = nodes_df['Gender'].map({'Male': 0, 'Female': 1})
nodes_df.to_csv('../data/classification_with_id2.csv', index=False)
```
