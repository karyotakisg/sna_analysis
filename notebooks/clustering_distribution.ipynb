```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
nodes = pd.read_csv('../data/nodes.csv')
nodes.columns
```

```python
df2 = nodes.groupby('clustering').size().reset_index(name='counts')
figure = plt.figure(figsize=(7, 4))
sns.scatterplot(x='clustering', y='counts', data=df2, s=100)
plt.savefig('../figures/clustering.png')
```

```python
figure = plt.figure(figsize=(7, 4))
plt.yscale('log')
plt.ylabel('Logarithmic Counts')
plt.xlabel('Clustering Coefficient')
sns.histplot(nodes['clustering'])
plt.savefig('../figures/clustering_hist.png')
```
